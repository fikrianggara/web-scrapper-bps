```python
import os
import pandas as pd
import io

import google.auth
import random
import math
from Google import Create_Service
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
from googleapiclient.http import MediaIoBaseDownload
```

```python
rekeningDf = pd.read_csv('rekening_petugas_clean.csv', sep=';')
```

```python
splittedDf = rekeningDf.head()
splittedDf.iloc[0:5]
```

```python
CLIENT_SECRET_FILE = 'credentials.json'
API_NAME='drive'
API_VERSION='v3'
SCOPES=['https://www.googleapis.com/auth/drive']

service = Create_Service(CLIENT_SECRET_FILE, API_NAME, API_VERSION, SCOPES)

```

```python
rootDir = "rekening"
count=1
total = len(rekeningDf.index)

for index, row in rekeningDf.iterrows():

    url = row['Foto Buku Rekening Aktif']
    id = url.split('=')[1]
    # create the output directory, the downloaded file should be put into individual folder
    outputDir = row['nama']+'-'+str(row['Wilayah Kerja (Kecamatan)'])+'-'+str(row['Nomor Rekening Aktif'])
    path=os.path.join(rootDir, outputDir)
    
    # check if the path directory is exist, 
    # if exist, create random id in the end of directory name 
    if(os.path.exists(path)):
        path = path+'('+str(random.randint(1,100))+')'
        
    os.makedirs(path)

    filename=os.path.join(path, 'buku_rekening.jpg')

    # get file using api created before
    request = service.files().get_media(fileId=id)
    fh = io.BytesIO()
    downloader = MediaIoBaseDownload(fd=fh, request=request)

    done = False

    while not done:
        status, done = downloader.next_chunk()
        # print('Download progress {0}'.format(status.progress()*100))

    fh.seek(0)

    # save downloaded file to the directory previously defined
    with open(filename, 'wb') as f:
        f.write(fh.read())
        f.close()
    progress=(count/total)*100
    progressPercent = math.ceil(progress*100)/100
    print('Overall progress {0}% ({1}/{2})'.format(progressPercent, count, total))
    count+=1
```
